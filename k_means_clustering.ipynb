<h1>The task is to implement k-means clustering for a dataset - named Iris which exists in the sklearn.datsets. The KMeans method from sklearn.cluster is used to fit the data in cluster. The number of cluster is picked based on the silhouette score and the number of cluster that resulted in the maximum score is found by iterating over different values in a reasonable range. </h1>


```python
from sklearn import datasets
import pandas as pd
from sklearn.cluster import KMeans
```

```python
from sklearn.metrics import silhouette_score
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
iris_ds = datasets.load_iris()
```

```python
df_iris = pd.DataFrame(iris_ds.data, columns=iris_ds.feature_names)
df_iris.head()
```

```python
df_iris = df_iris.rename(columns={'sepal length (cm)': 'SEPAL_LENGTH', 'sepal width (cm)': 'SEPAL_WIDTH', 'petal length (cm)': 'PETAL_LENGTH', 'petal width (cm)': 'PETAL_WIDTH'})
df_iris.head()
```

```python
df_iris.shape
```

```python
max_score = float('-inf')
num_cluster = -1
for k in range(2, 50):
  kmeans = KMeans(init="random", n_clusters=k, n_init=8, random_state=42)
  sil_score = silhouette_score(df_iris,kmeans.fit_predict(df_iris))
  if sil_score > max_score: num_cluster = k
  max_score = max(max_score,sil_score)
print(f"The maximum silhouette score: {max_score}\nCluster number with this score: {num_cluster}")
```

**The following appraoch shows the normalization on the dataset wouldn't change the score in this particular example. the standard normalization method is used.**

```python
for cols in df_iris:
  df_iris[cols] = scaler.fit_transform(df_iris[[cols]])
```

```python
for k in range(2, 50):
  kmeans = KMeans(init="random", n_clusters=k, n_init=8, random_state=42)
  sil_score = silhouette_score(df_iris,kmeans.fit_predict(df_iris))
  if sil_score > max_score: num_cluster = k
  max_score = max(max_score,sil_score)
print(f"The maximum silhouette score: {max_score}\nCluster number with this score: {num_cluster}")
```
